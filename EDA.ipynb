```python
# 1. Create a set of all unique body types
all_body_types = set()
data1['Treatment Body Type Titles'].str.split(', ').apply(all_body_types.update)

# 2. Create indicator columns for each body type (one-hot encoding)
for body_type in all_body_types:
    data1[f'{body_type}'] = data1['Treatment Body Type Titles'].str.contains(body_type).astype(int)
```

```python
data1['Face'] = data1['Face'] | data1['Face ']  # Bitwise OR for element-wise comparison

# Drop the old columns
data1.drop(['Face '], axis=1, inplace=True)
# Convert all string values in the DataFrame to uppercase
data1 = data1.applymap(lambda x: x.upper() if isinstance(x, str) else x)

data1.head(5)
```

```python
filtered_df = df[df['Face'] == 1]

# Extract individual values from the 'Categories' column
# Split values by ',' and flatten the resulting lists
Catalog_list_Face = filtered_df['Treatment Catalog Titles'].str.split(',').explode().unique()

print("Unique individual values where Flag is 1:")
print(Catalog_list_Face)
```

```python
# Substring to count
substring = "Botox"

# Count occurrences of the substring in the list
print('BOTOX',sum(s.count("BOTOX") for s in Catalog_list_Face))
print('WRINKLE',sum(s.count("WRINKLE") for s in Catalog_list_Face))
```

```python
from collections import Counter

# Split the strings into words and count occurrences
all_words = " ".join(Catalog_list_Face).split()  # Combine all strings and split into words
word_counts = Counter(all_words)  # Count occurrences of each word

# Sort by count in descending order
sorted_word_counts = word_counts.most_common()

print("Word occurrences in descending order:")
for word, count in sorted_word_counts:
    print(f"{word}: {count}")
```
